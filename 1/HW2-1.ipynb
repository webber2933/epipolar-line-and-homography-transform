```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# function of img show
def imshow(image):
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    else:
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
        
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
# read point from txt file
def read_point(file):
    f = open(file)
    temp = f.read().splitlines()
    
    number = len(temp) - 1
    point = np.ones((number, 3))
    
    for i in range(len(temp)):
        if i == 0:
            continue
        x = float(temp[i].split(' ')[0])
        y = float(temp[i].split(' ')[1])
        point[i-1,0] = x
        point[i-1,1] = y
    
    return point
```

```python
point1 = read_point('./pt_2D_1.txt')
point2 = read_point('./pt_2D_2.txt')
print(point1)
```

```python
# Implement the linear least squares eight point algorithm and report the returned fundamental matrix.
def linear_least_square(point1,point2):
    
    num = len(point1)
    
    A = np.zeros((num, 9))
    
    # construct the point correspondences matrix
    for i in range(num):
        u1 = point1[i][0]
        v1 = point1[i][1]
        u2 = point2[i][0]
        v2 = point2[i][1]
        A[i] = np.array([u1*u2, u1*v2, u1, v1*u2, v1*v2, v1, u2, v2, 1])
        
    # compute the SVD of A
    U, D, VT = np.linalg.svd(A, full_matrices=True)
    # take the rightmost column of V which corresponds to the smallest singular value
    f = VT[-1, :]
    # reshape f into a 3*3 matrix
    f = np.reshape(f, (3, 3))
    
    # enforce the rank-two constraint
    U, D, VT = np.linalg.svd(f, full_matrices=True)
    D_rank2 = np.zeros((3, 3))
    D_rank2[0][0] = D[0]
    D_rank2[1][1] = D[1]
    fundamental = np.dot(U, np.dot(D_rank2, VT))

    return fundamental
```

```python
fundamental_matrix = linear_least_square(point1,point2)
print(fundamental_matrix)
```

```python
# Implement the normalized eight point algorithm and report the returned fundamental matrix.
def normalized_eight_point_algorithm(point1,point2):
    
    num = point1.shape[0]
    
    point1_euclidean = point1[:, 0:2]
    point2_euclidean = point2[:, 0:2]
    
    mean1 = np.mean(point1_euclidean, axis=0)
    mean2 = np.mean(point2_euclidean, axis=0)
    
    point1_zero_mean = point1_euclidean - mean1
    point2_zero_mean = point2_euclidean - mean2
    
    total_dist = 0
    for i in point1_zero_mean:
        dist = np.sqrt(i[0]**2 + i[1]**2)
        total_dist += dist
    avarage_dist = total_dist / num
    scale1 = np.sqrt(2) / avarage_dist
    
    
    total_dist = 0
    for i in point2_zero_mean:
        dist = np.sqrt(i[0]**2 + i[1]**2)
        total_dist += dist
    avarage_dist = total_dist / num
    scale2 = np.sqrt(2) / avarage_dist
    
    T1 = np.array([[scale1, 0, -mean1[0] * scale1],
                   [0, scale1, -mean1[1] * scale1],
                   [0, 0, 1]])

    T2 = np.array([[scale2, 0, -mean2[0] * scale2],
                   [0, scale2, -mean2[1] * scale2],
                   [0, 0, 1]])
    
    normalized_pt1 = T1.dot(point1.T).T
    normalized_pt2 = T2.dot(point2.T).T
    fundamental = linear_least_square(normalized_pt1, normalized_pt2)
    fundamental = np.dot(np.dot(T1.T,fundamental),T2)
    return fundamental
```

```python
normalized_fundamental_matrix = normalized_eight_point_algorithm(point1,point2)
print(normalized_fundamental_matrix)
```

```python
# draw epipolar line
def draw_epipolar_line_and_cal_avg_dist(point1,point2,img1,img2,F):
    num = point1.shape[0]
    
    
    # the epipolar line on image 1, line1[0] = a, line1[1] = b, line1[2] = c, => ax+by+c = 0
    line1 = F.dot(point2.T)
    total_dist1 = 0
    avg_dist1 = 0
    for i in range(num):
        
        # draw line with (0,-c/b) and (W,-(aW + c)/b) on image1
        x1,y1,x2,y2 = 0,int(-line1[2][i]/line1[1][i]),img1.shape[1],int(-(line1[2][i]+line1[0][i]*img1.shape[1])/line1[1][i])
        cv2.line(img1,(x1,y1),(x2,y2),(0, 0, 255),1)
        
        # draw point on image1
        x1,y1 = int(point1[i][0]),int(point1[i][1])
        cv2.circle(img1,(x1,y1),2,(255,0,0),2)
        
        # cal the dist from point to line
        total_dist1 += np.abs(point1[i][0]*line1[0][i] + point1[i][1]*line1[1][i] + line1[2][i]) / np.sqrt(line1[0][i]**2 + line1[1][i]**2)
        
    avg_dist1 = total_dist1 / num
    print("the avarage dist : ",avg_dist1)
    imshow(img1)
    
    # the epipolar line on image 2, line1[0] = a, line1[1] = b, line1[2] = c, => ax+by+c = 0
    line2 = F.T.dot(point1.T)
    total_dist2 = 0
    avg_dist2 = 0
    for i in range(num):
        
        # draw line with (0,-c/b) and (W,-(aW + c)/b) on image2
        x1,y1,x2,y2 = 0,int(-line2[2][i]/line2[1][i]),img2.shape[1],int(-(line2[2][i]+line2[0][i]*img2.shape[1])/line2[1][i])
        cv2.line(img2,(x1,y1),(x2,y2),(0, 0, 255),1)
        
        # draw point on image2
        x1,y1 = int(point2[i][0]),int(point2[i][1])
        cv2.circle(img2,(x1,y1),2,(255,0,0),2)
        
        # cal the dist from point to line
        total_dist2 += np.abs(point2[i][0]*line2[0][i] + point2[i][1]*line2[1][i] + line2[2][i]) / np.sqrt(line2[0][i]**2 + line2[1][i]**2)
        
    avg_dist2 = total_dist2 / num
    print("the avarage dist : ",avg_dist2)
    imshow(img2)
    
    return img1,img2
```

```python
a_img1 = cv2.imread('./image1.jpg')
a_img2 = cv2.imread('./image2.jpg')
a_img1,a_img2 = draw_epipolar_line_and_cal_avg_dist(point1,point2,a_img1,a_img2,fundamental_matrix)
cv2.imwrite('./output/a_img1.jpg',a_img1)
cv2.imwrite('./output/a_img2.jpg',a_img2)
```
